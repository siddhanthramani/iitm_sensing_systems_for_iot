### Task 2: Compute and plot location errors using haversine distance [15 points] <a id='task2'></a>
Calculate the mean location coordinates for all three locations. Assume those three locations as the `groundtruth`. Compute the error for each location fix sample, which is the distance between the location sample and the groundtruth location. You must use haversine distance ([Wiki](https://en.wikipedia.org/wiki/Haversine_formula)). Plot the CDF (cumulative distribution function) of these errors for the three locations. Comment on the variance of the three distributions.


```python
# Imports
import pandas as pd
import numpy as np
from math import cos
```

```python
# Opening each log file and reading them
with open("./log_data/outdoor_gnss_log.txt", "r") as my_file:
    log_data_outdoor = my_file.read()
with open("./log_data/gateway_gnss_log.txt", "r") as my_file:
    log_data_gateway = my_file.read()
with open("./log_data/indoor_gnss_log.txt", "r") as my_file:
    log_data_indoor = my_file.read()
```

```python
# Parsing the fix logs
def parse_fix_log(log_line):
    list_log_line = log_line.split(",")
    temp_dict = {}
    # getting the lat and long values from appropriate index
    temp_dict["lat_deg"] = float(list_log_line[2])
    temp_dict["long_deg"] = float(list_log_line[3])
    flag = 0
    # ensuring that the lat and long values are not null
    for key, val in temp_dict.items():
        if not val:
            flag = 1
    # returning value if lat and long present
    if not flag:
        return temp_dict

# Parsing status logs
def parse_status_log(log_line):
    list_log_line = log_line.split(",")
    temp_dict = {}
    # getting the constellation type, svid, cn0, azimuth, elevation and used in fix values from appropriate index
    temp_dict["constellation_type"] = float(list_log_line[4])
    temp_dict["svid"] = float(list_log_line[5])
    temp_dict["cn0"] = float(list_log_line[7])
    temp_dict["azimuth_deg"] = float(list_log_line[8])
    temp_dict["elevation_deg"] = float(list_log_line[9])
    temp_dict["used_in_fix"] = float(list_log_line[10])
    flag = 0
    # If not beloning to GPS or not used in fix, ignore datapoint
    if int(list_log_line[4])!=1 or int(list_log_line[10]==0):
        flag = 1
    # If any value empty, ignore datapoint
    for key, val in temp_dict.items():
        if not val:
            flag = 1
    if not flag:
        return temp_dict
```

```python
# A main parser which which returns a dataframe with fix datapoints and a dataframe with status datapoints
def get_df_fix_and_df_status(log_data):    
    list_fix_logs = []
    list_status_logs = []
    list_log_data = log_data.split("\n")

    # iterating through each line
    for log_line in list_log_data:
        # If according to Fix specs, parse the log as a fix log
        if "Fix" in log_line and "GPS" in log_line:
            parsed_log = parse_fix_log(log_line)
            if parsed_log:
                list_fix_logs.append(parsed_log)
        # If according to Status specs, parse the log as a status log
        elif "Status" in log_line and "Svid" not in log_line:
            parsed_log = parse_status_log(log_line)
            if parsed_log:
                list_status_logs.append(parsed_log)

    # Create the fix datapoint
    df_fix = pd.DataFrame(list_fix_logs)
    # Convert deg columns to rad columns
    df_fix["lat_rad"] = np.deg2rad(df_fix['lat_deg'])
    df_fix["long_rad"] = np.deg2rad(df_fix['long_deg'])

    # Create the status dataframe
    df_status = pd.DataFrame(list_status_logs)
    # Convert the deg columns to rad columns
    df_status["azimuth_rad"] = np.deg2rad(df_status['azimuth_deg'])
    df_status["elevation_rad"] = np.deg2rad(df_status['elevation_deg'])
    return df_fix, df_status
```

```python
# Get the df_fix and df_status for all three log files
df_fix_outdoor, df_status_outdoor = get_df_fix_and_df_status(log_data_outdoor)
df_fix_gateway, df_status_gateway = get_df_fix_and_df_status(log_data_gateway)
df_fix_indoor, df_status_indoor = get_df_fix_and_df_status(log_data_indoor)
```

```python
# Getting the mean latitude and longitude for radian columns
def get_mean_lat_long_rad(df):
    mean_lat_rad = df["lat_rad"].mean()
    mean_long_rad = df["long_rad"].mean()
    return mean_lat_rad, mean_long_rad

# Radius of earth in kms for haversine error formula
r = 6371

# Calculating the haversine error
def get_haversine_error(df):
    # Gets mean values
    mean_lat_rad, mean_long_rad = get_mean_lat_long_rad(df)
    # Implementing the haversine formula (vectorially) and returning h_error
    a = np.sin((mean_lat_rad - df["lat_rad"])/2)**2 + np.cos(df["lat_rad"]) * cos(mean_lat_rad) * np.sin((mean_long_rad - df["long_rad"])/2)**2
    c = 2 * np.arcsin(np.sqrt(a)) 
    df["h_error"] = c * r
```

```python
# Calling the haversine error function for outdoor log file
mean_lat_outdoor, mean_long_outdoor = get_mean_lat_long_rad(df_fix_outdoor)
get_haversine_error(df_fix_outdoor)
print("Outdoor log - Mean lat : {} rad, Mean long : {} rad".format(mean_lat_outdoor, mean_long_outdoor))
df_fix_outdoor
```

```python
# Calling the haversine error function for gateway log file
mean_lat_gateway, mean_long_gateway = get_mean_lat_long_rad(df_fix_gateway)
get_haversine_error(df_fix_gateway)
print("Gateway log - Mean lat : {} rad, Mean long : {} rad".format(mean_lat_gateway, mean_long_gateway))
df_fix_gateway
```

```python
# Calling the haversine error function for indoor log file
mean_lat_indoor, mean_long_indoor = get_mean_lat_long_rad(df_fix_indoor)
get_haversine_error(df_fix_indoor)
print("Indoor log - Mean lat : {} rad, Mean long : {} rad".format(mean_lat_indoor, mean_long_indoor))
df_fix_indoor
```

```python
# Helper function to reset a variable
def reset_temp_dict():
    temp_dict = {}
    temp_dict["svid"] = []
    temp_dict["used_in_fix"] = []
    temp_dict["cn0"] = []
    temp_dict["count_svid_in_fix"] = 0
    return temp_dict

# A parser which combines fix and corresponding status log
def get_df_fix_status(log_data):
    list_fix_status_logs = []

    list_log_data = log_data.split("\n") 
    temp_dict = reset_temp_dict()

    for index in range(len(list_log_data)):
        log_line = list_log_data[index]
        list_log_line = log_line.split(",")
        # If status of a particular fix
        if "Status" in log_line and "Svid" not in log_line:
            # If constellation type is GPS and it has a fix value
            if int(list_log_line[4]) == 1 and list_log_line[5] and list_log_line[10]:
                # Append important info to list
                temp_dict["svid"].append(int(list_log_line[5]))
                temp_dict["used_in_fix"].append(int(list_log_line[10]))
                temp_dict["cn0"].append(float(list_log_line[7]))
                # If it has been used in fix, update count
                if int(list_log_line[10]):
                    temp_dict["count_svid_in_fix"] += 1
        elif "Fix" in log_line and "GPS" in log_line:
            # Get the lat and long values from fix log
            temp_dict["lat_deg"] = float(list_log_line[2])
            temp_dict["long_deg"] = float(list_log_line[3])
            list_fix_status_logs.append(temp_dict)
            temp_dict = reset_temp_dict()
    
    # Create the final log and calculate lat and long in radians
    df_fix_status = pd.DataFrame(list_fix_status_logs)
    df_fix_status["lat_rad"] = np.deg2rad(df_fix_status['lat_deg'])
    df_fix_status["long_rad"] = np.deg2rad(df_fix_status['long_deg'])


    return df_fix_status
```
